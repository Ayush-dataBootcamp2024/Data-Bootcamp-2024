```python
monotremes = ["platypus", "echidna"]
marsupials = ["kangaroo", "koala", "wombat"]
placentals = ["human", "dog", "cat", "whale", "elephant"]
 
species = input("Enter the species of the mammal: ").lower()
 
if species in monotremes:
    category = "Monotreme"
elif species in marsupials:
    category = "Marsupial"
elif species in placentals:
    category = "Placental"
else:
    category = "Unknown category"
 
print(f"The species {species} belongs to the category: {category}")
```

```python
def tempconvert():
    print("This function would convert degree C to degree F and vice versa")
    temp = float(input("Enter the value of temperature: "))
    unit = input("Enter the unit (C or F)")
    if unit.upper() == "C":
        f=(temp*9/5)+32
        return print(f"The temperature of {temp}{unit} is {f}F")
    elif unit.upper() =="F":
        c=(temp-32)*5/9
        return print(f"The temperature of {temp}{unit} is {f}C")
    else:
        return print("Invalid input, Either enter C for celsius or F for fahrenheit")
```

```python
def tempconvert():
    print("This function would convert degree C to degree F and vice versa")
    temp = float(input("Enter the value of temperature: "))
    unit = input("Enter the unit (C or F)")
    if unit.upper() == "C":
        fah=(temp*9/5)+32
        return print(f"The temperature of {temp} {unit.upper()} is {fah} F")
    elif unit.upper() =="F":
        c=(temp-32)*5/9
        return print(f"The temperature of {temp} {unit.upper()} is {c} C")
    else:
        return print("Invalid input, Either enter C for celsius or F for fahrenheit")
```
